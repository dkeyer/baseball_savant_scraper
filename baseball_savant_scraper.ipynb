```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
import re
```

```python
url = 'https://baseballsavant.mlb.com/savant-player/673357'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

table = soup.find('table', {'id': 'runValues'})

csv_file = open('luis_robert_run_value_2023.csv', 'w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Year','Pitch Type', 'Run Value', 'RV/100'])

for row in table.find_all('tr')[0:]:
    columns = row.find_all('td')
    year = columns[0].text.strip()
    pitch_type = columns[1].text.strip()
    run_value = columns[4].text.strip()
    rv_100 = columns[3].text.strip()
    csv_writer.writerow([year, pitch_type, run_value, rv_100])

csv_file.close()
```

```python
url = 'https://baseballsavant.mlb.com/leaderboard/custom?year=2023&type=batter&filter=&sort=1&sortDir=desc&min=10&selections=xba,xslg,xwoba,xobp,xiso,exit_velocity_avg,launch_angle_avg,sweet_spot_percent,barrel_batted_rate,&chart=false&x=xba&y=xba&r=no&chartType=beeswarm'
response = requests.get(url)
html_content = response.text

pattern = re.compile(r'"player_id":"([^"]+)"')
matches = pattern.findall(html_content)

player_identifiers = []

for match in matches:
    player_id = match
    player_identifiers.append({'Player ID': player_id})

# Print the player identifiers
for player in player_identifiers:
    print(f"Player ID: {player['Player ID']}")

```
